```python
import cv2
import numpy as np
from keras.models import load_model
import winsound  # For alert sound (Windows only)
```

```python
# Load trained model
model = load_model("face_mask_model.h5")

# Load Haar Cascade for face detection
face_detector = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")

# Start webcam
cap = cv2.VideoCapture(0)

# Optional: increase webcam resolution
cap.set(3, 640)
cap.set(4, 480)
```

```python

while True:
    ret, frame = cap.read()
    if not ret:
        print("вҡ пёҸ Camera not accessible!")
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect face
    faces = face_detector.detectMultiScale(gray, scaleFactor=1.2, minNeighbors=4)

    for (x, y, w, h) in faces:
        face_img = frame[y:y+h, x:x+w]

        try:
            resized = cv2.resize(face_img, (128, 128))
        except:
            continue  # Skip face if resize fails

        normalized = resized / 255.0
        reshaped = np.reshape(normalized, (1, 128, 128, 3))

        prediction = model.predict(reshaped)[0][0]

        label = "MASK" if prediction < 0.5 else "NO MASK"
        color = (0, 255, 0) if label == "MASK" else (0, 0, 255)

        # Warning sound only when no mask
        if label == "NO MASK":
            winsound.Beep(1000, 200)  # freq, duration

        cv2.putText(frame, label, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX,
                    0.8, color, 2)
        cv2.rectangle(frame, (x, y), (x + w, y + h), color, 2)

    cv2.imshow("рҹҳ· Live Face Mask Detector - Press Q to Exit", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
